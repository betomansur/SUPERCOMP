```python
import matplotlib.pyplot as plt
#3.12.0
```

```python
# 30
# limitado: 1.7
# bufado: 0.45

# 50
# limitado: 2.86
# bufado: 0.75

# 100
# limitado: 5.74
# bufado: 1.53
```

```python
x = [30,50,100]
limitado = [1.7,2.86,5.74]
bufado = [0.45,0.75, 1.53]

limitado = plt.plot(x, limitado, marker='o', label="Limitado")
bufado = plt.plot(x, bufado, marker='o',label = "Bufado")


plt.title('Gráfico de Linha')
plt.xlabel('Numero')
plt.ylabel('Tempo (s)')

plt.legend()


plt.show()
```

- **sinfo:** Verifica o estado e disponibilidade dos nós e partições no cluster.
- **squeue:** Monitora a fila de jobs, permitindo ver quais jobs estão em execução ou na fila de espera.
- **sprio:** Exibe a prioridade dos jobs, ajudando a entender a ordem de execução.
- **srun:** Submete jobs interativos ou em tempo real.
- **sstat:** Monitora o uso de recursos dos jobs em execução.
- **strigger:** Configura gatilhos para automação baseada em eventos no SLURM.
