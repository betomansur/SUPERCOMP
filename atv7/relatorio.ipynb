```python
import matplotlib.pyplot as plt

tempos_1_heuristica_1 = [0.00000533, 0.00000319, 0.00000295, 0.00000285, 0.00000286]
tempos_2_heuristica_1 = [0.00000356, 0.00000349, 0.00000345, 0.00000358, 0.00000346]
tempos_3_heuristica_1 = [0.00000417, 0.00000325, 0.00000326, 0.00000332, 0.00000334]
tempos_4_heuristica_1 = [0.00000249, 0.00000238, 0.00000242, 0.00000263, 0.00000240]

tempos_1_heuristica_2 = [0.00001655, 0.00001245, 0.00001191, 0.00001195, 0.00001209]
tempos_2_heuristica_2 = [0.00001433, 0.00001388, 0.00001408, 0.00001393, 0.00001392]
tempos_3_heuristica_2 = [0.00001324, 0.00001339, 0.00001771, 0.00001293, 0.00001308]
tempos_4_heuristica_2 = [0.00001032, 0.00001063, 0.00001060, 0.00001039, 0.00001048]


media_heuristica_1 = [
    sum(tempos_1_heuristica_1) / len(tempos_1_heuristica_1),
    sum(tempos_2_heuristica_1) / len(tempos_2_heuristica_1),
    sum(tempos_3_heuristica_1) / len(tempos_3_heuristica_1),
    sum(tempos_4_heuristica_1) / len(tempos_4_heuristica_1)
]

media_heuristica_2 = [
    sum(tempos_1_heuristica_2) / len(tempos_1_heuristica_2),
    sum(tempos_2_heuristica_2) / len(tempos_2_heuristica_2),
    sum(tempos_3_heuristica_2) / len(tempos_3_heuristica_2),
    sum(tempos_4_heuristica_2) / len(tempos_4_heuristica_2)
]

plt.figure(figsize=(10, 6))
conjuntos = ['Conjunto 1', 'Conjunto 2', 'Conjunto 3', 'Conjunto 4']

plt.plot(conjuntos, media_heuristica_1, marker='o', linestyle='-', label='Embaralhamento')
plt.plot(conjuntos, media_heuristica_2, marker='o', linestyle='-', label='Probabilidade')


plt.xlabel('Conjuntos')
plt.ylabel('Tempo Médio de Execução (segundos)')
plt.title('Comparação de Tempos Médios de Execução entre as Heurísticas')
plt.legend()

plt.tight_layout()
plt.show()
```
